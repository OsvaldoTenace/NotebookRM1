#1. Esfuerzo y Deformacion Normal Axial
## 1.1 Esfuerzo
El **esfuerzo** se mide en términos de fuerza por unidad de área y se representa con el símbolo griego **σ** (sigma) cuyas unidades son de fuerza sobre unidad de área.

Supongamos que los esfuerzos que actúan sobre la sección transversal (como se muestra en la figura 1.1b) se distribuyen de manera uniforme en toda el área. En este caso, la resultante de estos esfuerzos debe ser igual a la magnitud del esfuerzo por unidad de área de la sección transversal A de la barra, es decir, P = σA. Por lo tanto, obtenemos la siguiente expresión para la magnitud de los esfuerzos:

$$ σ={\frac{P}{A}} \quad (1.1)  $$

Cuando la barra se somete a fuerzas de tracción representadas por P, los esfuerzos que se generan son de tipo tensión. Si se invierte la dirección de las fuerzas de manera que la barra es comprimida, entonces se producen esfuerzos de compresión.

Cuando es necesario establecer una convención de signos para los esfuerzos normales, es común definir los esfuerzos de tensión como positivos y los esfuerzos de compresión como negativos.

La ecuación 1.1 es aplicable únicamente cuando el esfuerzo está distribuido de manera uniforme en toda la sección transversal de la barra. **Esta condición se cumple cuando la fuerza axial P actúa en el centroide del área de la sección transversal**, como se demuestra en la sección 1.3. ##1.2. Deformación Unitaria Normal:

En términos generales, el alargamiento de un segmento es igual a su longitud dividida por la longitud total L y multiplicada por el alargamiento δ, como se observa en la figura 1.1a. Por lo tanto, una longitud unitaria de la barra experimentará un alargamiento igual a 1/L multiplicado por δ. Esta cantidad se conoce como alargamiento por unidad de longitud, o **deformación unitaria**, y se representa con la letra griega ε (épsilon). Podemos notar que la deformación unitaria se calcula mediante la siguiente ecuación:

$$ ε={\frac{δ}{L}} \quad (1.2)   $$

Cuando la barra se encuentra en tensión, la deformación unitaria se denomina **deformación unitaria por tensión**, lo que representa un alargamiento o estiramiento del material. Por otro lado, cuando la barra está en compresión, la deformación unitaria se llama **deformación unitaria por compresión**, y la barra se acorta. En general, la deformación unitaria por tensión se considera positiva, mientras que la deformación unitaria por compresión se considera negativa. La deformación unitaria, representada por ε, se denomina **deformación unitaria normal** debido a que está relacionada con los esfuerzos normales en la estructura. Dado que la deformación unitaria normal es la relación entre dos longitudes, es una **cantidad adimensional**, lo que significa que carece de unidades de medida. Por lo tanto, la deformación unitaria se expresa simplemente como un número, sin depender de ningún sistema de unidades específico.



##1.3. Línea de Acción de las fuerzas Axiales para una distribución uniforme del esfuerzo:

En todo el análisis previo sobre el esfuerzo y la deformación unitaria en una barra prismática, asumimos que **el esfuerzo normal σ estaba distribuido de manera uniforme** a lo largo de su sección transversal. Ahora, demostraremos que esta condición se cumple cuando la línea de acción de las fuerzas axiales pasa a través del centroide del área de la sección transversal de la barra.

Consideremos una barra prismática con una sección transversal de forma arbitraria que está sometida a fuerzas axiales P que generan esfuerzos σ distribuidos de manera uniforme. Supongamos que el punto **p$_1$** en la sección transversal es donde la línea de acción de las fuerzas intersecta la sección, como se muestra en la figura 1.2. Si trazamos un conjunto de ejes xy en el plano de la sección transversal y representamos las coordenadas del punto p1 como **$\bar{x}$** y **$\bar{y}$**, para determinar esas coordenadas, observamos que los momentos  **M$_x$** y **M$_y$** generados por la fuerza P con respecto a los ejes x e y, respectivamente, deben ser iguales a los momentos correspondientes de los esfuerzos distribuidos uniformemente.

$$ M_x = P\bar{y} \quad M_y = P\bar{x} \quad (1.3,1.4)$$



```python
#@title Figura 1.2. Distribución uniforme de esfuerzos en una barra prismática
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Rectangle

# Definir el rango de valores para el parámetro t
t = np.linspace(0, 2*np.pi, 1000)

# Definir las ecuaciones paramétricas para x(t) e y(t)
x = np.cos(t) + np.sin(0.5*t)+1.5
y = np.sin(t)+0.75

# Ángulo de rotación en radianes
angulo_rotacion = np.radians(30)  # Convertir 30 grados a radianes

# Matriz de rotación en el plano
matriz_rotacion = np.array([[np.cos(angulo_rotacion), -np.sin(angulo_rotacion)],
                            [np.sin(angulo_rotacion), np.cos(angulo_rotacion)]])

# Aplicar la rotación a las coordenadas x e y
coordenadas = np.vstack((x, y))  # Apilar las coordenadas x e y
coordenadas_rotadas = np.dot(matriz_rotacion, coordenadas)  # Aplicar la rotación

x_rotado, y_rotado = coordenadas_rotadas

# Graficar la curva paramétrica rotada
plt.figure(figsize=(6, 6))
plt.plot(x_rotado, y_rotado, label='Curva paramétrica rotada', color='black', alpha = 0.7)
# Rellenar el área interior de la curva cerrada
plt.fill(x_rotado, y_rotado, color='c', alpha=0.1)  # Relleno en color cian
plt.axis('equal')  # Para mantener la proporción en los ejes x e y

#establecemos limites
plt.xlim(0, 3)
plt.ylim(0, 3)

# Calcular el centroide de la curva centrada
centroid_x = np.mean(x_rotado)
centroid_y = np.mean(y_rotado)

#graficamos el punto del centroide
plt.scatter(centroid_x,centroid_y,color='black')

# Graficar las cotas
plt.quiver(centroid_x/2+0.25, 2.25, -1, 0, angles='xy', scale_units='xy', scale=1)
plt.quiver(centroid_x/2-0.25, 2.25, 1, 0, angles='xy', scale_units='xy', scale=1)

plt.quiver(2.25, centroid_y/2+0.13 , 0, -1, angles='xy', scale_units='xy', scale=1)
plt.quiver(2.25, centroid_y/2 -0.13 , 0, 1, angles='xy', scale_units='xy', scale=1)

recta_x = np.linspace(centroid_x,2.25,1000)
constante_y = np.full_like(recta_x, centroid_y)

plt.plot(recta_x,constante_y , color='black', alpha = 0.5)

recta_y = np.linspace(centroid_y,2.25,1000)
constante_x = np.full_like(recta_y, centroid_x)

plt.plot(constante_x ,recta_y, color='black', alpha = 0.5)

#dibujamos el cuadrado
square = Rectangle((1.25, 1.25), 0.1, 0.1, facecolor = 'c', edgecolor='black')
plt.gca().add_patch(square)

# Graficar las cotas
plt.quiver(1.3/2, 1.3, 0.6, 0, angles='xy', scale_units='xy', scale=1)
plt.quiver(1.3/2, 1.3, -0.65, 0, angles='xy', scale_units='xy', scale=1)
plt.quiver(1.3, 1.3/2, 0, 0.6, angles='xy', scale_units='xy', scale=1)
plt.quiver(1.3, 1.3/2, 0, -0.65, angles='xy', scale_units='xy', scale=1)

recta_x = np.linspace(centroid_x,2.25,1000)
constante_y = np.full_like(recta_x, centroid_y)

plt.plot(recta_x,constante_y , color='black', alpha = 0.5)

recta_y = np.linspace(centroid_y,2.25,1000)
constante_x = np.full_like(recta_y, centroid_x)

plt.plot(constante_x ,recta_y, color='black', alpha = 0.5)

#ponemos los caracteres
plt.quiver(1, 2.75, 0.2, -0.3, angles='xy', scale_units='xy', scale=1, width = 0.003)
plt.text(0.9, 2.8, 'A', fontsize=15, color='black')

plt.text(centroid_x, centroid_y-0.15, 'p$_1$', fontsize=15, color='black')
plt.text(1.25, 1.4, 'dA', fontsize=15, color='black')
plt.text(centroid_x/2-0.05,2.3, 'x\u0304', fontsize=15, color='black')
plt.text(1.25/2,1.35, 'x', fontsize=15, color='black')
plt.text(2.3,centroid_y/2-0.05, 'y\u0304', fontsize=15, color='black')
plt.text(1.35,1.25/2, 'y', fontsize=15, color='black')

plt.xticks([])  # Oculta los valores del eje X
plt.yticks([])  # Oculta los valores del eje Y

# Configurar flecha en el eje X
plt.annotate('', xy=(3, 0.01), xytext=(0, 0),
             arrowprops=dict(arrowstyle='-|>', linewidth=1.5, color='black',alpha = 0.8))

# Configurar flecha en el eje Y
plt.annotate('', xy=(0,2.99), xytext=(0, 0),
             arrowprops=dict(arrowstyle='-|>', linewidth=1.5, color='black',alpha = 0.8))

plt.xlabel('X')
plt.ylabel('Y')

plt.show()


```

Los momentos de los esfuerzos distribuidos se obtienen al realizar una integral sobre el área de la sección transversal **A**. La fuerza diferencial que actúa sobre un elemento de área **dA** (como se muestra en la figura 1.2) es igual a **σdA**. Los momentos de esta fuerza elemental con respecto a los ejes x e y son **σ$\cdot$ydA** y **-σ$\cdot$xdA**, respectivamente, donde **x** e **y** representan las coordenadas del elemento **dA**. Los momentos totales se obtienen integrando sobre el área de la sección transversal de la siguiente manera:

$$ M_x =\intσ\cdot y  \, dA \quad M_y =\int-σ\cdot x  \, dA \quad (1.5,1.6)$$

Igualando las ecuaciones 1.3 y 1.5, y las ecuaciones 1.4 y 1.6:

$$ \bar{y} = \frac{\int y  \, dA}{A} \quad  \bar{x} = \frac{\int x  \, dA}{A} \quad  (1.7,1.8)$$

Dado que los esfuerzos σ están distribuidos uniformemente, sabemos que son constantes en toda el área **A** de la sección transversal, y, por lo tanto, pueden ser factorizados fuera de los signos de integración. Además, sabemos que *σ* es igual a *P/A*. Estas ecuaciones son idénticas a las que definen las coordenadas del centroide de un área. Por lo tanto, hemos llegado a una conclusión significativa: para lograr una tensión o compresión uniforme en una barra prismática, la fuerza axial debe aplicarse en el centroide del área de la sección transversal.




##1.3 Diagrama de esfuerzo-deformacion unitaria:
Tras realizar un ensayo de tensión o compresión y determinar tanto el esfuerzo como la deformación unitaria para varias magnitudes de carga, es posible construir un gráfico que muestra el esfuerzo en función de la deformación unitaria. Este **diagrama esfuerzo-deformación unitaria** es una característica distintiva del material específico sometido a ensayo y proporciona información valiosa acerca de sus propiedades mecánicas y su tipo de comportamiento.

Observando la figura 1.3, el diagrama comienza con una línea recta que se extiende desde el origen O hasta el punto A, indicando que en esta región inicial, la relación entre el esfuerzo y la deformación unitaria no sólo es lineal sino también proporcional. Más allá del punto A, la proporcionalidad entre el esfuerzo y la deformación unitaria ya no se mantiene, y es por eso que al esfuerzo en el punto A se le denomina **límite de proporcionalidad**.

La pendiente de la línea recta de O a A se denomina *módulo de elasticidad*. Debido a que la pendiente tiene unidades de esfuerzo dividido entre la deformación unitaria, el módulo de elasticidad tiene las mismas unidades que el esfuerzo.

Con un aumento en el esfuerzo más allá del límite de proporcionalidad, la deformación unitaria comienza a aumentar a un ritmo más acelerado con cada incremento en el esfuerzo. Como resultado, la curva esfuerzo-deformación unitaria muestra una pendiente cada vez menor, y en el punto B, la curva se vuelve horizontal. Desde este punto en adelante, la muestra de ensayo experimenta un alargamiento significativo sin un aumento sustancial en la fuerza de tensión (de B a C). Este fenómeno se denomina **fluencia** del material, y el punto B se conoce como **punto de fluencia**. El esfuerzo correspondiente se denomina **esfuerzo de fluencia**.

Después de las considerables deformaciones unitarias que se producen durante la fluencia en la región BC, se comienza a experimentar un proceso de **endurecimiento por deformación**. Durante este endurecimiento por deformación, el material experimenta cambios en su estructura cristalina, lo que resulta en una mayor resistencia del material a una deformación adicional. La elongación de la muestra de ensayo en esta región requiere un aumento en la carga de tensión, y, por lo tanto, el diagrama esfuerzo-deformación unitaria muestra una pendiente positiva de C a D. Finalmente, la carga alcanza su valor máximo y el esfuerzo correspondiente, en el punto D, se denomina **esfuerzo último**. Un alargamiento adicional de la barra en realidad se acompaña de una reducción en la carga, y la fractura ocurre finalmente en un punto como el E en la figura 1.3.

##1.4 Ley de Hooke
La relación lineal que ocurre en la región OA de la figura 1.3 entre el esfuerzo y la deformación unitaria para un material en tension o compresion simple se describe con la siguiente ecuación:

$$ σ = E\cdotε \quad (1.9)$$

Los términos **σ** y **ε** ya fueron descritos anteriormente como el esfuerzo y la deformación unitaria, respectivamente. **E** es una constante de proporcionalidad, que corresponde a la **pendiente de la recta OA** (mencionado en la sección 1.3), denominada módulo de elasticidad y es propia de cada material. La ecuación 1.9 es conocida como la **Ley de Hooke**. Esta es, de hecho, una versión muy limitada de la Ley de Hooke, ya que solo se relaciona con los esfuerzos longitudinales y las deformaciones unitarias desarrolladas en situaciones de tensión o compresión simple de la barra, es decir, en un solo eje del elemento.


##1.5. Cambios de Longitud en Elementos cargados Axialmente
###1.5.1. Resortes

El alargamiento de un resorte se ilustra en la figura 1.4, donde en la parte superior de la figura se muestra un resorte en su **longitud natural L** (también conocida como longitud sin esfuerzo, longitud relajada o longitud libre), y en la parte inferior de la figura se representan los efectos al aplicar una carga de tensión. Debido a la acción de la fuerza **P**, el resorte se estira una cantidad **δ** y su longitud final es **L + δ**. Si el material del resorte es linealmente elástico, la carga y el alargamiento serán proporcionales, siguiendo las siguientes ecuaciones:

$$ P = k\cdotδ \quad ; \quad δ =  f\cdot P \quad  (1.10)$$

La constante **k** se denomina **rigidez** del resorte y se define como la fuerza necesaria para producir un alargamiento unitario, es decir, k = P/δ . De manera similar, la constante f se conoce como flexibilidad y se define como el alargamiento producido por una carga de valor unitario, es decir, f = δ/P.


```python
#@title Figura 1.4. Alargamiento de un resorte cargado axialmente.
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import gridspec

fig = plt.figure(figsize=(9, 12))

# Definir la disposición de subgráficos utilizando GridSpec
gs = gridspec.GridSpec(1, 2, width_ratios=[1, 1])

# Crear los subplots utilizando GridSpec
ax1 = plt.subplot(gs[0], projection='3d')
ax2 = plt.subplot(gs[1], projection='3d')

# Ecuación paramétrica del resorte
phi = np.linspace(0, 2 * np.pi, 1000)
z = 1 * np.cos(6 * phi)
y = 1 * np.sin(6 * phi)

for ax in [ax1, ax2]:
    if ax == ax1:
      x = np.linspace(0, 1, 1000)
      c=1
    else:
      x = np.linspace(0, 3, 1000)
      c=3
    # Graficar las fuerzas
    fuerza1_inicio = [0, 0, 0]
    fuerza1_direccion = [-1, 0, 0]
    longitud1 = 1

    fuerza2_inicio = [c, 0, 0]
    fuerza2_direccion = [1, 0, 0]
    longitud2 = 1

    ax.quiver(*fuerza1_inicio, *fuerza1_direccion, length=longitud1, color='r', arrow_length_ratio=0.45, linewidth=1.5, capstyle='round')
    ax.quiver(*fuerza2_inicio, *fuerza2_direccion, length=longitud2, color='r', arrow_length_ratio=0.45, linewidth=1.5, capstyle='round')
    ax.text(-1,0,0.5,'P',fontsize= 20, color = 'red')
    ax.text(c+1,0,0.5,'P',fontsize= 20, color ='red')

    # Ajustar el rango de los ejes
    ax.set_xlim([np.min(x) - 1, np.max(x) + 1])
    ax.set_ylim([-2, 2])
    ax.set_zlim([-2, 2])

    # Graficamos las cotas
    cota1_inicio = [c/2, -c/2, -1.5]
    cota1_direccion1 = [-1, 0, 0]
    cota1_direccion2 = [1, 0, 0]
    longitud_cota1 = c/2

    ax.quiver(*cota1_inicio, *cota1_direccion1, length=longitud_cota1, color='black', arrow_length_ratio=0.2, linewidth=1, capstyle='round')
    ax.quiver(*cota1_inicio, *cota1_direccion2, length=longitud_cota1, color='black', arrow_length_ratio=0.2, linewidth=1, capstyle='round')

    # Graficar el resorte en ambos subplots
    ax.plot(x, y, z, linewidth=3, color='c')
    ax.axis('off')

ax1.text(0.7, -c/2, -1, 'L', fontsize=12, va='top', ha='center')
ax2.text(1, -c/2, -2, 'L+δ', fontsize=12, va='top', ha='center')

plt.tight_layout()
plt.show()
```

###1.5.2. Barras Prismaticas.
Una barra prismática es un elemento estructural que posee un eje longitudinal recto y una sección transversal constante a lo largo de toda su longitud. La figura 1.1 (a) muestra el alargamiento δ de una barra prismática cuando se somete a una carga de tensión P. Si la carga actúa en el centroide de la sección transversal extrema, el esfuerzo normal σ uniforme en secciones transversales alejadas de los extremos se describe mediante la ecuación 1.1. Además, si la barra está hecha de un material homogéneo, la deformación unitaria axial ε sigue la ecuación 1.2.

También supongamos que el material es linealmente elástico, lo cual significa que obedece la Ley de Hooke. En este caso, los esfuerzos y las deformaciones unitarias están relacionados por la ecuación 1.9. Al combinar estas relaciones fundamentales, obtenemos la siguiente ecuación para el alargamiento de la barra.

$$ δ={\frac{P\cdot L}{E\cdot A}} \quad (1.11)$$

La rigidez y la flexibilidad de una barra prismática se definen de la misma manera que se hizo para un resorte. La **rigidez** es la fuerza necesaria para producir un alargamiento unitario, o **P/δ**, y la **flexibilidad** es su alargamiento debido a una carga unitaria, o **δ/P**. Por lo tanto, a partir de la ecuación 1.11, observamos que la **rigidez** y la **flexibilidad** de una barra prismática son, respectivamente:

$$ k={\frac{E\cdot A}{L}} \quad  f={\frac{L}{E\cdot A}} \quad (1.12,1.13)$$

##1.6. Efectos Termicos.
Los cambios de temperatura causan la expansión o contracción de los materiales estructurales, lo que conduce a deformaciones térmicas y esfuerzos térmicos.Un ejemplo sencillo de dilatación térmica se ilustra en la figura 1.5, donde el bloque de material no está restringido y, por lo tanto, tiene la libertad de expandirse. Cuando se calienta el bloque, cada elemento del material experimenta deformaciones térmicas en todas las direcciones, y como resultado, las dimensiones del bloque aumentan.

Para la mayoría de los materiales estructurales, la deformación unitaria térmica T es directamente proporcional al cambio de temperatura ΔT. En otras palabras, se puede expresar como:

$$ ε_T =α\cdotΔT \quad (1.14)$$

Donde **α** es el *coeficiente de expansión térmica* del material. Para establecer una convención de signos, para las deformaciones unitarias térmicas, se supone que la dilatación e positiva y la contracción negativa

Para establecer una relación entre el desplazamiento **δ** y la **variación de temperatura ΔT**,  partimos de la ecuación 1.2, combinando con la ecuación 1.14, dando como resultado:

$$ δ = ε_T\cdot L=α\cdotΔT\cdot L \quad (1.15)$$

```python
#Resolucion del ejercicio
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import gridspec
from matplotlib.patches import Rectangle

##Datos tensor CD
E_1 = 72*10**9 #pa
D_1 = 4*10**(-3) #m
L_1 = 0.4 #m
s_1 = 200*10**6 #pa

##Datos tensor EF
E_2=45*10**9 #pa
D_2=3*10**(-3) #m
L_2= 0.3 #m
s_2=175*10**6 #pa

##Calculo de Areas
A_1 = np.pi*D_1**2/4
A_2 = np.pi*D_2**2/4

##Sumatoria de momentos respecto a A T_1 +2*T_2 = 3P
##Se plantean las ecuaciones de las deformaciones de cada cable d_1 y d_2 con la ecuacion 1.11
##Compatibilidad de deformaciones (T_1*L_1)/(E_1*A_1)*2= (T_2*L_2)/(E_2*A_2)

##Sistema de ecuaciones
coeficientes = np.array([[1,2],[2*(L_1)/(E_1*A_1),-(L_2)/(E_2*A_2)]])
constantes = np.array([3,0])

##solucion de las fuerzas T_1 y T_2
T_1 , T_2 = np.linalg.solve(coeficientes, constantes)

##Analisis de falla
if T_1/A_1 <= T_2/A_2: #tensiones a partir de ecuacion 1.1
  P = s_2/(T_2/A_2)
else:
  P = s_1/(T_1/A_1)

T_1*=P
T_2*=P

##Declaracion de variables de fuerzas en tensores y de deformaciones
d_1=(T_1*L_1)/(E_1*A_1)
d_2=(T_2*L_2)/(E_2*A_2)

#Reaccion vertical en A
R_av = T_1+T_2-P

#Reaccion horizontal en A
R_ah = 0

##Graficos (DCL) y Analisis de deformaciones
fig = plt.figure(figsize=(20, 5))
##Definir la disposición de subgráficos utilizando GridSpec
gs = gridspec.GridSpec(1, 2, width_ratios=[1, 1])

##subplot del DCL
###dibujo de la barra
ax1 = plt.subplot(gs[0])
barraAB = Rectangle((0, 0.2),1, 0.05, facecolor = 'c', edgecolor='black')
plt.gca().add_patch(barraAB)
ax1.set_xlim([ - 0.2,1.2])
ax1.set_ylim([ 0,0.45])

#fuerzas dibujadas
plt.quiver(0, 0.1, 0, 0.1, angles='xy', scale_units='xy', scale=1, color='grey')
plt.quiver(-0.15, 0.225, 0.15, 0, angles='xy', scale_units='xy', scale=1, color='grey')
plt.quiver(1/3, 0.25, 0, 0.1, angles='xy', scale_units='xy', scale=1, color='green')
plt.quiver(2/3, 0.25, 0, 0.1, angles='xy', scale_units='xy', scale=1, color='blue')
plt.quiver(1, 0.2, 0, -0.1, angles='xy', scale_units='xy', scale=1, color='red')

##texto
plt.text(0.0,0.07, f'{round(R_av,2)} kg', fontsize=15, color='grey', ha='center', va='center')
plt.text(-0.1,0.25, f'{round(R_ah,2)} kg', fontsize=15, color='grey', ha='center', va='center')
plt.text(1/3,0.37, f'{round(T_1,2)} kg', fontsize=15, color='green', ha='center', va='center')
plt.text(2/3,0.37, f'{round(T_2,2)} kg', fontsize=15, color='blue', ha='center', va='center')
plt.text(3/3,0.07, f'{round(P,2)} kg', fontsize=15, color='red', ha='center', va='center')

cotas = [(0,'A'),(1/3,'D'),(2/3,'F'),(1,'B')]

for i,j in cotas:

  plt.text(i+0.02,0.225, f'{j}', fontsize=15, color='black', ha='center', va='center')
  plt.scatter(i, 0.225,color='black')

##Compatibilidad de deformaciones
ax2 = plt.subplot(gs[1])

##grafico del eje de la barra
eje_barra = np.linspace(0,1,100)
ubicacion_y = np.full_like(eje_barra, 0.225)
ax2.plot(eje_barra,ubicacion_y,color = 'black')

##grafico del eje de la barra rotado
ax2.plot(eje_barra,eje_barra*(-d_1/(1/3))+0.225,color = 'black')

##grafico de desplazamientos
k=0
for i,j in [(d_1,'green'),(d_2,'blue'),(d_1*3,'red')]:
  k+=1/3
  desplazamiento = np.linspace(0.225,0.225-i,100)
  ubicacion_x = np.full_like(desplazamiento, k)
  ax2.plot(ubicacion_x,desplazamiento, color = j)
  plt.text(k+0.05,0.2248, f'{round(i*100,4)} cm', fontsize=10, color=j, ha='center', va='center')

##texto
for i,j in cotas:

  plt.text(i,0.22510, f'{j}', fontsize=15, color='black', ha='center', va='center')
  plt.scatter(i, 0.225,color='black')

##ajuste del limites del grafico
ax2.set_xlim([ - 0.1,1.1])
ax2.set_ylim([ 0.225-d_1*4,0.225+d_1*4])

##ocultar ejes
ax1.axis('off')
ax2.axis('off')

plt.tight_layout()
plt.show()
```

```python
#Resolucion del ejercicio
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import gridspec
from matplotlib.patches import Rectangle

D_T = 80 #°C
L= 1 #m
a= 1.2*10**(-5) #1/°C
E= 200*10**9 #pa
A= 5/100**2 #m^2

#Ecuacion de compatibilidad dR=-dT
dT=a*D_T*L #se aplica la ecuacion 1.15
dR=-dT

#Calculo de la Reaccion a partir de la ecuacion 1.11
R=E*A*dR/L

##Graficos (DCL) y Analisis de deformaciones
fig = plt.figure(figsize=(20, 8))

##Definir la disposición de subgráficos utilizando GridSpec
gs = gridspec.GridSpec(1, 3, width_ratios=[1, 1,1])

##subplots
ax1 = plt.subplot(gs[0])
ax2 = plt.subplot(gs[1])
ax3 = plt.subplot(gs[2])

#Lista de subplots
subplots = [ax1,ax2,ax3]

#Dibujo de la barra original en cada subplot
for ax in subplots:
  barraAB = Rectangle((-0.2, L/4),0.4, L, facecolor = 'c', edgecolor='black')#grafico de la longitud original de la barra AB
  ax.add_patch(barraAB)
  #limites de los graficos
  ax.set_xlim([ - 1,1])
  ax.set_ylim([ 0,1.5*L])
  #empotramiento inferior
  empotramiento =  Rectangle((-0.25,0),0.5, L/4, facecolor = 'gray', edgecolor='black')
  ax.add_patch(empotramiento)
  #cota de la longitud original
  datos_x= np.linspace(-1,1,1000)
  cte_y = np.full_like(datos_x,L+L/4)
  ax.plot(datos_x, cte_y, color='black', linestyle = '--')
  ax.plot(datos_x, cte_y-L, color='black', linestyle = '--')
  #puntos A y B
  ax.scatter(0, L+L/4,color='black')
  ax.text(0.1, L+L/4.3, 'A', fontsize=15, color='black', ha='center', va='center')
  ax.scatter(0, L/4,color='black')
  ax.text(0.1, L/3.7, 'B', fontsize=15, color='black', ha='center', va='center')
  #Elementos exclusivos del primer grafico
  if ax == ax1:
    #emportramiento superior
    empotramiento =  Rectangle((-0.25,L+L/4),0.5, L/4, facecolor = 'gray', edgecolor='black')
    ax.add_patch(empotramiento)
    #Reacciones superiores e inferiores. Condicion de ajuste en caso de cambio de sentido en la variacion de temperatura
    if D_T > 0:
      ax.quiver(0, L+L/2.5, 0, -L/6.666667, angles='xy', scale_units='xy', scale=1, color='red')
      ax.quiver(0,L/10, 0, L/6.666667, angles='xy', scale_units='xy', scale=1, color='red')
    elif D_T<0:
      ax.quiver(0, L+L/4, 0, L/6.666667, angles='xy', scale_units='xy', scale=1, color='red')
      ax.quiver(0,L/4, 0, -L/6.666667, angles='xy', scale_units='xy', scale=1, color='red')
    #textos grafico 1
    ax.text(-0.42,(1.5*L)/2, f'\u0394T={D_T}°C', fontsize=20, color='orange', ha='center', va='center')
    ax.text(0,L/20, f'R={round(R,2)} kg', fontsize=12, color='red', ha='center', va='center')
    ax.text(0,1.5*L-L/20, f'R={round(R,2)} kg', fontsize=12, color='red', ha='center', va='center')
  #Elementos graficos exclusivos del segundo grafico
  if ax == ax2:
    #texto de temperatura
    ax.text(-0.42,(1.5*L)/2, f'\u0394T={D_T}°C', fontsize=20, color='orange', ha='center', va='center')
    #rectangulo de la deformacion de temperatura
    alargamiento_T= Rectangle((-0.2, L+L/4),0.4, dT*100, facecolor = 'orange', edgecolor='black')
    ax.add_patch(alargamiento_T)
    #texto de la deformacion
    ax.text(0,L+L/4+dT*100/2, 'δ' + '\u1D40' +' =' f'{round(dT*100,4)} cm', fontsize=10, color='black', ha='center', va='center')
  #Elementos graficos exclusivos del tercer grafico
  if ax == ax3:
    #Reaccion superior con condicion de ajuste por cambio de sentido
    if D_T > 0:
      ax.quiver(0, L+L/2.5, 0, -L/6.666667, angles='xy', scale_units='xy', scale=1, color='red')
    elif D_T<0:
      ax.quiver(0, L+L/4, 0, L/6.666667, angles='xy', scale_units='xy', scale=1, color='red')
    #texto de valor de la reaccion
    ax.text(0,1.5*L-L/20, f'R={round(R,2)} kg', fontsize=12, color='red', ha='center', va='center')
    #deformacion producida por la reaccion
    acortamiento_R = Rectangle((-0.2, L+L/4+dR*100),0.4, -dR*100, facecolor = 'red', edgecolor='black')
    ax.add_patch(acortamiento_R)
    ax.text(0,L+L/4+dR*100/2, 'δ' + '\u1D3F' +' =' f'{round(dR*100,4)} cm', fontsize=10, color='black', ha='center', va='center')
  ax.axis('off')#apagamos los ejes

plt.tight_layout()
plt.show()

```

Una barra cónica $AB$ con sección transversal circular sólida y longitud $L$ está empotrada en el extremo $B$ y sometida a una carga de tensión $P$ en su
extremo libre $A$. Los diámetros de la barra en los extremos $A$ y $B$ son $d_A$ y $d_B$, respectivamente.

Determine el alargamiento de la barra debido a la carga P, suponga que el ángulo de conicidad es pequeño

DATOS:

$L = 3 m$

$P = 20 tn$

$d_A = 2 cm$

$d_B = 6 cm$

$E= 2.1\cdot10^6 kg/cm^2$

image.png

```python
#Resolucion del ejercicio
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import gridspec
from matplotlib.patches import Polygon

#Datos
L=300 #cm
P= 20000 #kg
d_a= 2 #cm
d_b = 6 #cm
E =2.1*10**6 #kg/cm2

datos_x = np.linspace(0,L,1000)

#funciond de variacion del radio
def funcion_radio(d_1,d_2):
  return (d_2-d_1)/(2*L)*datos_x+d_1/2

#funcion de variacion del area
def funcion_area(d_1,d_2):
  return np.pi*funcion_radio(d_1,d_2)**2

#funcion constante de la fuerza N
funcion_P = np.full_like(datos_x,P)

#funcion constante del modulo E
funcion_E = np.full_like(datos_x,E)

#calculo de la deformacion por integral definida de 0 a L de [N(x)*dx]/[E(x)A(x)]
deformacion = np.trapz(funcion_P/(funcion_E*funcion_area(d_a,d_b)),x=datos_x)

##Graficos (DCL) y Analisis de deformaciones
fig = plt.figure(figsize=(15, 8))

##Definir la disposición de subgráficos utilizando GridSpec
gs = gridspec.GridSpec(3, 1, height_ratios=[1, 1,1])
##subplots
ax1 = plt.subplot(gs[0]) #diagrama de fuerza normal
ax2 = plt.subplot(gs[1]) #grafico de la figura

#Grafico de la funcion del diagrama de fuerza normal
ax1.plot(datos_x,funcion_P,color ='blue')
#graficar rectas verticales DFN
for i in [0,L]:
  ax1.plot(np.full_like(funcion_P,i),np.linspace(0,P,1000), color = 'blue')
#Limites del grafico DFN
ax1.set_xlim([ -2,L+2])
ax1.set_ylim([ 0,P+ P/abs(P)*2000])
#labels DFN
ax1.set_xlabel('X')
ax1.set_ylabel('N(x)')
ax1.set_title('Diagrama de Fuerza Normal.')

#establecer las coordenadas de los vertices de la figura
vertices = [(0, -1), (0, 1), (300, 3), (300, -3)]
v_empotramiento = [(300,-3.5),(300,3.5),(302,3.5),(302,-3.5)]
vertices_def = [(0-deformacion*30, -1), (0-deformacion*30, 1), (300, 3), (300, -3)]

# Crear el trapecio usando la clase Polygon
trapecio = Polygon(vertices, closed=True, edgecolor='black', facecolor='c')
ax2.add_patch(trapecio)
#figura del empotramiento
empotramiento = Polygon(v_empotramiento, closed=True, edgecolor='black', facecolor='grey')
ax2.add_patch(empotramiento)
#figura de la barra deformada
trapecio_def = Polygon(vertices_def, closed=True, edgecolor = 'red',facecolor='none', linestyle = '--',linewidth=3)
ax2.add_patch(trapecio_def)
ax2.legend([trapecio, trapecio_def], ['Antes de deformarse', 'Después de deformarse'], loc='lower left')

#cotas
ax2.plot(np.linspace(0,-deformacion*30,100),np.full_like(np.linspace(0,-deformacion,100),2), color = 'black')
ax2.plot(np.full_like(np.linspace(1.9,2.1,50),-deformacion*30),np.linspace(1.7,2.3,50), color = 'black')
ax2.plot(np.full_like(np.linspace(1.9,2.1,50),0),np.linspace(1.7,2.3,50), color = 'black')
#texto de cotas
ax2.text(-deformacion*30/2,2.5, f'{round(deformacion,3)} cm', fontsize=8, color='black', ha='center', va='center')

ax2.set_title('Antes y Despues de la Deformacion.')

ax2.set_xlim([ -20,L+2.5])
ax2.set_ylim([ -3.2,3.2])
ax2.axis('off')

plt.tight_layout()
plt.show()
```
